# Dataset Overview

## Dataset Description

The **DataCo Supply Chain Dataset** is a structured retail supply chain dataset that contains transactional, operational, and customer-related data generated throughout the order fulfillment process. The dataset combines information from multiple business functions, including customer management, product management, sales, order processing, shipping, delivery, and financial performance.

Each record represents an individual order item transaction, enabling detailed analysis of supply chain operations at a transactional level.

The dataset provides a comprehensive view of business activities, making it suitable for evaluating operational efficiency, customer purchasing behavior, sales performance, profitability, logistics performance, and regional market trends.

## Key Performance Indicators (KPIs)

| KPI | Description | Business Importance |
|------|-------------|--------------------|
| Total Sales | Total revenue generated from all orders | Measures business growth |
| Total Profit | Total profit earned from completed orders | Measures financial performance |
| Profit Margin | Profit earned relative to sales | Evaluates profitability |
| Average Order Value | Average revenue generated per order | Measures customer purchasing behaviour |
| Order Volume | Total number of orders | Measures business demand |
| Late Delivery Rate | Percentage of delayed deliveries | Evaluates logistics efficiency |
| On-Time Delivery Rate | Percentage of orders delivered on schedule | Measures delivery performance |
| Average Shipping Time | Average actual shipping duration | Measures operational efficiency |
| Sales by Market | Revenue generated across different markets | Identifies high-performing markets |
| Profit by Category | Profit contribution by product category | Supports product portfolio decisions |

## Initial Business Expectations
